```python
import pandas as pd

df=pd.read_csv('customer_shopping_behavior.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower()  # removes spaces, lowercases everything
df.columns
```

```python

df['review rating'] = df.groupby('category')['review rating'].transform(lambda x: x.fillna(x.mean()))
df.isnull().sum()
```

```python
# rename 
df = df.rename(columns={'purchase amount (usd)': 'purchase_amount'})
df.columns
```

```python
# creare new age group 

labels = ['young-adult', 'adult','middle-age','senior']
df["age group"] = pd.qcut(df['age'], q=4, labels=labels)
df.head(1)
```

```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase frequency days'] = df['frequency of purchases'].map(frequency_mapping)

df.head(1)
```
